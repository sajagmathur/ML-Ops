```python
#figure out how much data is null --> Calculate what percentage of values are null
train_df_null = round((train_df.isnull().sum())/train_df.shape[0],4)*100
train_df_null = train_df_null.sort_values(ascending = False).reset_index().rename(columns = {'index':'columns',0:'Percentage'})

#reset_index makes it a dataframe --> make df with more than 30 percent nulls
train_df_null_fin = train_df_null[train_df_null['Percentage']>=30]
train_df_null_fin
```

```python
#Remove the relevant columns

train_df_updated = train_df.drop(columns = train_df_null_fin['columns'])
#train_df.info()
```
