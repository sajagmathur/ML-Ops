```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained model (e.g., all-MiniLM-L6-v2 is fast and effective)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sentence to embed
sentence = "I Love MLOps"

# Generate the embedding
embedding = model.encode(sentence)

print("Vector embedding:")
print(embedding)
print("Shape:", embedding.shape)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

lm_name = "gpt2"  # Small language model

tokenizer = AutoTokenizer.from_pretrained(lm_name)
model = AutoModelForCausalLM.from_pretrained(lm_name)

prompt = "Tell me what is 2+2."

inputs = tokenizer(prompt, return_tensors="pt")

outputs = model.generate(
    inputs['input_ids'],
    max_length=200,               # 10,000 is very long; 200 is typical for explanation
    do_sample=True,
    temperature=0.8,
    pad_token_id=tokenizer.eos_token_id  # Avoid warning by explicitly setting pad_token_id
)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("\nGenerated Explanation:\n", generated_text)
```

```python
from google_genai import client

# Initialize the GenAI client (uses environment variable GOOGLE_APPLICATION_CREDENTIALS)
genai_client = client.GenAIClient()

# Select the Gemini model - e.g., "models/chat-bison-001" is the conversational Gemini model
model_name = "models/chat-bison-001"

prompt = "Explain the concept of MLOps in simple terms."

response = genai_client.generate_text(
    model=model_name,
    prompt=prompt,
    max_output_tokens=200,
    temperature=0.7,
)

print("\nGenerated Explanation:\n", response.text)
```
